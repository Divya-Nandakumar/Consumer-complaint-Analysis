```python
rus = RandomUnderSampler(random_state=42, replacement=False)# fit predictor and target variable 
# null values in target variable need to be eliminated 
x_rus1, y_rus1 = rus.fit_resample(x1, y1)

print('original dataset shape:', Counter(y1))
print('Resample dataset shape', Counter(y_rus1))
```

```python
final_response = pd.concat([x_rus1,y_rus1],axis=1)
```
